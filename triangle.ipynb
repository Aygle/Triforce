```python
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
import random


fig = plt.figure(figsize=(50,50))

ax = fig.add_subplot(111)

ax.axis('equal')

point1 = [0, 0]
point2 = [1, 0]
point3 = [0.5, 0.8]

x_values1 = [point1[0], point2[0]]
y_values1 = [point1[1], point2[1]]
x_values2 = [point3[0], point2[0]]
y_values2 = [point3[1], point2[1]]
x_values3 = [point1[0], point3[0]]
y_values3 = [point1[1], point3[1]]

ax.plot(x_values1, y_values1, "-k")
ax.plot(x_values2, y_values2, "-k")
ax.plot(x_values3, y_values3, "-k")

def pu(x , y, fig) :
    fig.plot(x, y,  ".k")

def area(x1,y1,x2,y2,x3,y3):
    return abs((x1*(y2-y3)+(x2*(y3-y1))+(x3*(y1-y2)))/2.0)

def calculerCoordonneesMilieu(A,B):
    x = (A[0]+B[0])/2   #calcul de l'abscisse
    y = (A[1]+B[1])/2   #calcul de l'ordonn√©e
    return [x,y]

for i in range(1) :
    x1 = np.random.random_sample() 
    y1 = np.random.random_sample()
    A=area(point1[0],point1[1],point2[0],point2[1],point3[0],point3[1])
    A1=area(point1[0],point1[1],x1,y1,point3[0],point3[1])
    A2=area(x1,y1,point2[0],point2[1],point3[0],point3[1])
    A3=area(point1[0],point1[1],point2[0],point2[1],x1,y1)

    while (A1)+(A2)+(A3)!=A :
        x1 = np.random.random_sample() 
        y1 = np.random.random_sample()
        A=area(point1[0],point1[1],point2[0],point2[1],point3[0],point3[1])
        A1=area(point1[0],point1[1],x1,y1,point3[0],point3[1])
        A2=area(x1,y1,point2[0],point2[1],point3[0],point3[1])
        A3=area(point1[0],point1[1],point2[0],point2[1],x1,y1)
    
    pu(x1 , y1 , ax)
    pts =[point1 , point2 , point3]
    for i in range(1000000) :
        c = random.choice(pts)
        [x1, y1 ] = calculerCoordonneesMilieu([x1 , y1], c)
        pu(x1 , y1 , ax)
```
